```python
import pandas as pd

# Load your Excel file (update path if needed)
df = pd.read_excel("GreenTrail Store Data.xlsx")

# Show first few rows
df.head()
```

```python
df['Sales Lift (%)'] = (
    (df['Weekly Sales During Promotion'] - df['Weekly Sales Before Promotion']) 
    / df['Weekly Sales Before Promotion']) * 100

df[['Store ID', 'Sales Lift (%)']].sort_values('Sales Lift (%)', ascending=False).head()
```

```python
import matplotlib.pyplot as plt

df['Sales Lift (%)'].plot(kind='hist', bins=10, title='Sales Lift Distribution')
plt.xlabel('Sales Lift (%)')
plt.ylabel('Number of Stores')
plt.show()
```
